```python
import numpy as np

def compute_eigenvectors(data, lag_time):
    n = len(data)

    # Covariance matrices
    cov = np.cov(data.T)
    cov_tau = np.cov(data[:-lag_time].T, data[lag_time:].T)[0:len(data[0]),len(data[0]):]  # update this line

    # Solve the generalized eigenvalue problem
    eigvals, eigvecs = np.linalg.eig(np.linalg.inv(cov) @ cov_tau)

    # Sort eigenvalues and eigenvectors by decreasing order
    sort_idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[sort_idx]
    eigvecs = eigvecs[:,sort_idx]
    
#     return eigvals
    return eigvecs
    
# test
data = np.random.normal(size=(100, 3))  # replace with your data
lag_time = 2  # replace with your lag time

eigvecs = compute_eigenvectors(data, lag_time)
# eigvals = compute_eigenvectors(data, lag_time)

print("Eigenvectors: ", eigvecs)
# print('vals',eigvals)
```
